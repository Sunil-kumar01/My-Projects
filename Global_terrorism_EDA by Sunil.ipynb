# *  Gangs involved in continuous activity.

```python
max_killevery_gname=[]
year=[]
g_name=[]
for i in after_droped_2:
  if i=='year':
    for j in range(2004,2018):
      year.append(j)
      attack=after_droped_2[after_droped_2['year']==j]
      max_killevery_gname.append(attack.gang_name.value_counts()[1])
      g_name.append(attack.gang_name.value_counts()[1:2].keys())
most_kill = pd.DataFrame({'year' : year,
                                'gname' : g_name,
                                'Number of Attacks' : max_killevery_gname })
most_kill['gname']=most_kill['gname'].apply(lambda x: x[0] )
plt.figure(figsize=(15,10))
plt.xlabel('Year')
plt.ylabel('Number of Attack')
plt.title('Year wise most active gang')
plt.plot(most_kill['year'],most_kill['Number of Attacks'])
for i in range(len(most_kill['year'])):
        plt.text(most_kill['year'][i],most_kill['Number of Attacks'][i],most_kill['gname'][i][0:12],rotation=70,fontsize=15)
```

**From this plot we can conclude few points**
* Most active gangs are Taliban,CPI-Maoist,ISIL
